## Sin Annealing

```python
x = []
n = 10000
for i in range(n):
    eps = 0.9 - (i/n)*0.01**(np.sin(i/4)/100) #Linear annealing from 8% to 1%
    x.append(eps)
plt.figure()
plt.plot(x)
plt.title("x")
```

## Linear Annealing

```python
x = []

for i in range(100):
    eps =  max(0.01, 0.08 - 0.01*(i/200))
    x.append(eps)
plt.figure()
plt.plot(x)
plt.title("x")
```
