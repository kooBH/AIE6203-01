## Linear Annealing

```python
x = []
n = 1000
start = 1
end = 0.1
decay = 1000

for step in range(n):
    if step >= decay:
        eps = end
    else :
        eps = (start* (decay - step) + end * step) / decay + 0.15*((n-step)/n)*np.cos(step/4)
    
    if eps > 1.0 :
        eps = 1.0
    
    x.append(eps)
plt.figure()
plt.plot(x)
plt.title("eps")
```
