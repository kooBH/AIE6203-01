```python
import gym
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
```

```python
env = gym.make("ALE/KungFuMaster-v5",obs_type="grayscale")
env = env.unwrapped
```

```python
state = env.reset()[0]
plt.ion()
fig= plt.figure(figsize=(8, 8))
im = plt.imshow(state,cmap='gray', vmin=0, vmax=255)

terminated = False
reward = 0.0
while not reward > 0.0 :
    while not terminated:
        clear_output(wait=True)
        state, reward, terminated,truncated,info = env.step(np.random.randint(13))
        if reward > 0.0 :
            print(reward)
            break
    
im.set_data(state)
display(fig)    
#clear_output(wait = True)
print(reward)
print(info)
```
